```
# 2)  WINE DATA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

wine_df = pd.read_csv("winequality_red.csv")
wine_df.head()
```

```
# bar plot
x = wine_df["fixed acidity"]
y = wine_df["residual sugar"]
plt.bar(x,y)
```

```
# heat maps
plt.subplots(figsize=(10,8))
ax = plt.axes()
ax.set_title("Wine Characteristic Correlation Heatmap (Reds)")
corr = wine_df.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
           cmap="Reds")
plt.show()
```

```
# line chart

columns = wine_df.columns.drop(['quality'])
# create x data
x_data = range(0, wine_df.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, wine_df[column])
# set title and legend
ax.set_title('Red Wine Dataset')
ax.legend()
```

```
# histogram
wine_df.plot.hist(subplots=True, layout=(6,6), figsize=(10, 10), bins=20)
```

```
# pairplots
sns.pairplot(wine_df)
```

```
# 3)  DIABETES DATA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

diabetes_df = pd.read_csv("diabetes.csv")
diabetes_df.head()
```

```
# histogram - data distribution of the features
diabetes_df.groupby('Outcome').hist(figsize=(9, 9))
```

```
# bar plot
x = diabetes_df["Age"]
y = diabetes_df["BMI"]
plt.bar(x,y)
```

```
# box plot

diabetes_df.plot.box(grid = "True")
```

```
# scatter plot
x = diabetes_df['Glucose']
y = diabetes_df['Insulin']
plt.scatter(x,y)
```

```
# heat maps
plt.subplots(figsize=(10,8))
ax = plt.axes()
ax.set_title("Diabetes characs")
corr = diabetes_df.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
           cmap="Reds")
plt.show()
```
